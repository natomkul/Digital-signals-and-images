```python
import numpy as np
import matplotlib.pyplot as plt

# Definiujemy oryginalnƒÖ funkcjƒô
def f(x):
    return np.sin(x)

# Definiujemy jƒÖdro konwolucji h(x)
def h(t):
    return np.where((t >= 0) & (t < 1), 1, 0)

# Tworzymy funkcjƒô konwolucji z jƒÖdrem h
def convolve_with_kernel(x_sample, y_sample, x_dense):
    y_dense = np.zeros_like(x_dense)
    dx = x_sample[1] - x_sample[0]  # Przyjmujemy sta≈Çy odstƒôp miƒôdzy pr√≥bkami

    # Dla ka≈ºdego punktu x w siatce interpolowanej
    for i, x in enumerate(x_dense):
        # Sumujemy warto≈õci y_sample * h(x - x_sample)
        y_dense[i] = np.sum(y_sample * h((x - x_sample)/dx))

    return y_dense

# Generujemy pr√≥bki funkcji w zakresie od -œÄ do œÄ
N = 20
x_sample = np.linspace(-np.pi, np.pi, N)  # 20 punkt√≥w pr√≥bki
y_sample = f(x_sample)

# Tworzymy gƒôstszƒÖ siatkƒô do interpolacji
x_dense = np.linspace(-np.pi, np.pi, 10*N)
y_dense = convolve_with_kernel(x_sample, y_sample, x_dense)

def mean_squared_error(y_true, y_pred):
    mse = np.mean((y_true - y_pred) ** 2)
    return mse

# Rysowanie oryginalnej funkcji i interpolacji
plt.figure(figsize=(10, 6))
plt.plot(x_dense, f(x_dense), label="Oryginalna funkcja sin(x)", color="blue")
plt.plot(x_dense, y_dense, label="Interpolacja za pomocƒÖ konwolucji", linestyle="--", color="red")
plt.scatter(x_sample, y_sample, color="green", marker="o", label="Punkty pr√≥bki")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title(f'Interpolacja funkcji sin(x) za pomocƒÖ konwolucji z jƒÖdrem h1(x) (MSE:{mean_squared_error(f(x_dense), y_dense)})')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from google.colab import drive
drive.mount('/content/drive')

x=io.imread("/content/drive/MyDrive/lena.jpg")

def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray

x=rgb2gray(x)


#k=pomniejszanie obrazu, s=rozmiar obrazu
def con(old_img, k):
  new_img=np.zeros((int(x.shape[0]//k), int(x.shape[1]//k)))
  for i in range(int(x.shape[0]//k)):
    for j in range(int(x.shape[1]//k)):
        block = old_img[i*k:(i+1)*k, j*k:(j+1)*k]
        pix = np.sum(block) / (k**2)
        new_img[i, j] = pix
  return new_img

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Oryginalny obraz")
plt.imshow(x, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Obraz po konwolucji")
plt.imshow(con(x,4), cmap='grey')
plt.show()
```
